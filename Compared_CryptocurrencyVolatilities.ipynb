# **Summary of Trends  Among the 24 Cryptocurrencies Included in this Survey**

The goal of ths study is to find suitable cryptocurrencies for use in buy-and-sell transactions and also for use as investment vehicles.
  
  We want to strike an acceptable balance between low volatility, low transaction fees, scalability. fast transaction speeds, privacy and security,  and interoperability.

As an initial phase in this study, we compare the actual price movements and price changes of cryptocurrencies which are frequently top listed in various of the criteria listed in the preceding paragraph.

There is a common trend to be seen among the 24 cryptocurrencies  included in this survey  : daily price percentage changes , (defined as (Price_today  -  Price_yesterday)/Price_yesterday ), tends to frequently oscillate between  increases and offsetting decreases so that average price percentage changes are much smaller than either average price percentage increases or average price percentage decreases. 
Price volatility as measured by annualized historic variances for cryptocurrency market returns for these various and / cryptocurrencies during the period 1/1/19 through 1/19/20  ranged between .72 and a high of 1.11 (for Aion). 
A number of cryptocurrencies were listed among top choices in multiple lists in multiple categories which this survey focused on, as is seen in upcoming slides  : These include Cardano, Litecoin, Monero, Dash, Zcash, Bitcoin Cash, EOS, XRP, Ethereum, and Zilliqa.
  Therefore, some balance, trade-off, may be is acceptable between making a  selection  with higher price volatility but which combines low transaction fees with other desirable criteria such as interoperability, scalability,and /  or privacy  and/ or high marketcaps and/ or fast transaction speeds or being top-listed as a better investment for other criteria.

```python

Insight 1: 
    SUMMARY: Bivariate Correlation which calculates Pearson's Correlation Co-efficients, showed
negligible correlation between cryptocurrency price percentage change per day, )which is the basic 
measure of price volatility), to any of the other attributes. 
However, the vrioius cryptocurrency daily prices, (Opening, Close, High and Low), were measured as
correlated to each other at 1, i.e. ,100 %, the maximum possible; and the prices were also 
strongly correlated to market volume at .56, ( .5 or above is considered to be a strong correlation),
and were very strongly correlated to marketcap at .95, as would be expected in view of definition of
marketcap

DETAILS:

(1) a Correlation Matrix which was generated by Python's .corr() function which uses Pearson's 
Correlation method ; and, directly below it,

(2) Barplot chart also generated by Python's .corr() function which presents the results listed in 
the Correlation matrix in an easier to comprehend, better readable format, especially for focusing 
on our primary attributes of interest

The fact that the matrix shows little or no correlation between cryptocurrency price percentage change 
per day, i.e., price volatility,  and any of the other variables, merely shows that there are
NO LINEAR relationships between cryptocurrency volatility and the other attributes 

KEY INSIGHT : Pearson Correlation Coefficient computation provided us valuable service as a process of
elimination, allowing us to skip other methods based on linear assumptions, in favor of other types of
analytic methods suitable in these circumstances : Correlation Matrix is often a preliminary step for
selecting whivh variables to include in applying other methods based on linear approximation such as 
linear regression; in this case, the results quickly indicated that using those methods would be a 
wast of time:  This Pearson Correlation Coefficient computation served the valauble purpose of quickly
moving on to apply other types of analytic methods suitable in these circumstances.
    
Daily Percentage Change in Price Showed negligible, if any, correlation with any of the other
parameters 

The Correlation Matrix below displays correlation coefficients for the top  12 cryptocurrencies 
in this survey, including the Top Five Cryptocurrencies for Investment while the others are listed
based their highest marketcaps. 

(High marketcap is  often deemed to be an indicator of investment potential).

The various daily prices, (Opening, Close, High and Low), were measured as correlated to each other 
at 1, i.e. ,100 %, the maximum possible.

The prices were also strongly correlated to market volume at .56, ( .5 or above is considered to be
a strong correlation), and were very strongly correlated to marketcap at .95, as would be expected 
in view of definition of marketcap.
Top 12 list rank/position had weaker correlations with the other parameters : .32 to daily prices,
 .21 to market volume, and .36 to marketcap.
However, percentage change in price per day, (which is the basic measure of price volatility) had the
weakest correlations with any of the other parameters, close to zero, although percentage change 
in daily closing price to changes in opening price, and in price highs, and  changes in price lows
per day. ranged from 100%,, the maximum, to the next strongest correlation of .99. )     

This indicates that price volatility could not be reliably predicted based on a linear regression
equation with the other parameters as the independent variables.     

    
```
